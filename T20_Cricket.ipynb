# Batting

```python
# eg:
a=[2,3]
b=[4,5]
a.extend(b)
a
```

```python
with open('C:/Users/SUHAS K/Downloads/Data Analytics Course/end_to_end_project/Data Analyst - PROJECT 1/PROJECT 1/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    for i in data:
        all_records.extend(i['battingSummary']) #getting the values of each dictionary key(battingSummary) using i pointing to each key & adding(extending) in all_records.
```

```python
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting.tail()
```

```python
#Adding a new column as "OUT/NOTOUT" & filled only when the batsman is "dismissal", nothing but a batsman is out if the dismissal column is having data for that batsman.
df_batting["Out/NotOut"] = df_batting.dismissal.apply(lambda x:"Out" if len(x)>0 else "Not Out")
```

```python
df_batting.head()
```

```python
#Using "Out/NotOut" column for analysis, so dropping a column "dismissal".
df_batting.drop(columns="dismissal", inplace=True) #inplace=true which means changes is keeping as it is.
df_batting.tail()
```

```python
#Finding the discrepency(special characters) in the batsman name & removing(eg: Kusal Mendisâ€,Vriitya Aravindâ€ ) - by spending time on analysis of data we will get.
df_batting['batsmanName']= df_batting['batsmanName'].apply(lambda x:x.replace('â€',''))
df_batting['batsmanName']= df_batting['batsmanName'].apply(lambda x:x.replace('\xa0',''))
df_batting.head(30)
```

```python
df_batting.head()
```

```python
#As their is no matching column b/w Batting & Match table, we have two columns(team1 & team2) in match table but we have bit different in batting table as column match(team1 vs team2). Here the thing is we need to find the match id which are same are not which the match happens b/w team1 vs team2 or team2 vs team1.
#eg:Namibia vs Srilanka : T20I # 1823, Srilanka vs Namibia : T20I # 1823..... so on like this we need to create dictionary.
match_id_dict = {}

for index,row in df_match.iterrows():                #iterrows - runs on each row 
    key1 = row['team1']+ " Vs " +row['team2']
    key2 = row['team2']+ " Vs " +row['team1']
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']
```

```python
match_id_dict #2-2 pairs were created for each match_id
```

```python
match_id_dict['Sri Lanka Vs Namibia']
```

```python
match_id_dict['Namibia Vs Sri Lanka']
```

```python
df_batting['match_id']=df_batting['match'].map(match_id_dict) #based on the match values the match_id will be picked & added init.
df_batting.head() #now we got common "match_id" for the batting table. 
```

```python
#As of now we have transformed the data by converting semi-structured(json) to structured format(tabular), cleaned the data by removing special characters, created a matching column by performing some analysis.
df_batting.shape
```

```python
#Converting this into csv format for creating visualization.
df_batting.to_csv("fact_batting_summary.csv", index=False)
```
